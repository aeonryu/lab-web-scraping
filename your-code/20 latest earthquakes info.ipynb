```python
# This is the url you will scrape in this exercise
url = 'https://www.emsc-csem.org/Earthquake/'

import requests
from bs4 import BeautifulSoup
import pandas as pd

html = requests.get(url).content
soup = BeautifulSoup(html, "lxml")
tabla = []
for a in soup.find_all('tr'):
    tabla.append(a.text.strip().split("\n")) 
#tabla

del tabla[0:14]
tabla
```

```python
del tabla[25:]
len(tabla)
tabla[0] #Ahora sí, cada renglón es una tabla
```

```python
import re
##re.split('[,\s]+', "a1 a2 a3, a4,a5")
#tabla[0] = re.split(r'[,\s]+', tabla[0][0])
#tabla[0]

for i in range(len(tabla)):
    tabla[i] = re.split(r'[,\s]+', tabla[i][0]) #regex para separar por comas o espacios
tabla
```

```python
df = pd.DataFrame(tabla)
df
```
